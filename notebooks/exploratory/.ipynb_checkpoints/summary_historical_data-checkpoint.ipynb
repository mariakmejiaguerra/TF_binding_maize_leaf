````python
import sys
import os
import pandas as pd #to read old xls and new xlsx files
````

````python
attribute_folder = 'GRIN PI Data'
attribute_path = "/data/Historical Data/{}/".format(attribute_folder)
for root, dirs, data_files in os.walk(attribute_path):
    data_files_xls = [filename for filename in data_files if filename.endswith(('.xls', '.xlsx'))]
    print("{}, number of data files: {}".format(attribute_folder, len(data_files_xls)))
    for filename in data_files_xls:
        if filename.endswith(('xls','xlsx')):
            xls_file = os.path.join(attribute_path,filename)
            print(filename)
            workbook = pd.ExcelFile(xls_file)
            for sheet in workbook.sheet_names:
                print("=" * 80)
                print('columns in sheet: {}'.format(sheet) )
                print("=" * 80)
                excel_data_df = pd.read_excel(workbook, sheet)
                excel_data_df.info()
                
````

````python
def read_historical_files(root_dir, attribute_dirs, read_attribute):
    count_files = 0
    count_sheetnames = 0
    for attribute in attribute_dirs:
        attribute_path = "{}/{}/".format(root_dir,
                                     attribute)
        if attribute != read_attribute:
            continue
        for root, years, files in os.walk(attribute_path):
            if years != []:
                years = sorted(years, key=lambda x: int(x))
                print("{}, number of years with data: {}".format(attribute,
                                                                 len(years)))
                for year in years:
                    attribute_year_path = "{}/{}/{}/".format(root_dir,
                                                             attribute,
                                                             year)
                    for root, dirs, data_files in os.walk(attribute_year_path):
                        n_data_files = len([filename for filename in data_files if filename.endswith(('.xls', '.xlsx'))])
                        #print("{}, number of data files: {}".format(year, n_data_files))
                        for filename in sorted(data_files):
                            file_sheets = 0
                            if filename.endswith(('.xls', '.xlsx')):
                                xls_file = os.path.join(attribute_year_path,filename)
                                workbook = pd.ExcelFile(xls_file) #the whole excel file
                                count_files = count_files + 1
                                file_sheets = len(workbook.sheet_names)
                                count_sheetnames = count_sheetnames + file_sheets
                            print("{}, total sheets: {}".format(filename, file_sheets))
        print("{}, total files: {}".format(attribute, count_sheetnames))
    return(count_files)
````

````python
historical_data_root_dir = '/data/Historical Data'
historical_data_dirs = ['Color Data',
                        'Cooking Data',
                        'Dry Weight Data',
                        'Insect Data']

to_explore_dir = 'Color Data'

read_historical_files(historical_data_root_dir, historical_data_dirs, to_explore_dir)
````

````python
to_explore_dir = 'Cooking Data'
read_historical_files(historical_data_root_dir, historical_data_dirs, to_explore_dir)
````

````python
to_explore_dir = 'Dry Weight Data'
read_historical_files(historical_data_root_dir, historical_data_dirs, to_explore_dir)
````

````python
to_explore_dir = 'Insect Data'
read_historical_files(historical_data_root_dir, historical_data_dirs, to_explore_dir)
````

````python
def describe_historical_files(root_dir, attribute_dirs, read_attribute):
    for attribute in attribute_dirs:
        attribute_path = "{}/{}/".format(root_dir,
                                     attribute)
        if attribute != read_attribute:
            continue
        for root, years, files in os.walk(attribute_path):
            if years != []:
                years = sorted(years, key=lambda x: int(x))
                for year in years:
                    attribute_year_path = "{}/{}/{}/".format(root_dir,
                                                             attribute,
                                                             year)
                    for root, dirs, data_files in os.walk(attribute_year_path):
                        n_data_files = len([filename for filename in data_files if filename.endswith(('.xls', '.xlsx'))])
                        for filename in sorted(data_files):
                            if filename.endswith(('.xls', '.xlsx')):
                                xls_file = os.path.join(attribute_year_path,filename)
                                workbook = pd.ExcelFile(xls_file) #the whole excel file
                                print("=" * 80)
                                print("year: {}".format(year))
                                print("file: {}".format(filename))
                                print("=" * 80)
                                for sheet in workbook.sheet_names:
                                    if "pivot" in sheet.lower() or sheet.startswith(('Sheet')):
                                        continue
                                    print("-" * 80)
                                    print('columns in sheet: {}'.format(sheet) )
                                    print("-" * 80)
                                    excel_data_df = pd.read_excel(workbook, sheet)
                                    excel_data_df.info()
````

````python
to_explore_dir = 'Color Data'
describe_historical_files(historical_data_root_dir, historical_data_dirs, to_explore_dir)
````
